```python
import pickle

def load_model(file_path):
    """
    Load a serialized model from a file using pickle.

    Parameters:
        file_path (str): Path to the serialized model file.

    Returns:
        object: The deserialized model.
    """
    with open(file_path, 'rb') as file:
        model = pickle.load(file)
    return model
```

```python
rf_model = load_model("rf_model_first.pkl")
rf_model
```
